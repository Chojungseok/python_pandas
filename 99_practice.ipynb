```python
df_copy.head(3)
```

```python
# Gentoo종의 flipper_length 평균
spec_gentoo = df_copy['species'] == 'Gentoo' # 'species'열에서 'Gentoo'가 True인 조건을 변수에 저장
df_copy.loc[spec_gentoo, 'flipper_length_mm'].mean() # 조건에 해당하는 데이터들의 'flipper_length_mm'열의 평균
```

```python
# Female Chinstrap종의 bill_length 높은순 정렬
## Female이면서 Chinstrap종인 데이터들의 bill_length데이터를 높은순(내림차순)정렬
spc_Chinstrap = df_copy['species'] == 'Chinstrap' # 종 들 중에 'Chinstrap'종 조건에 맞는 데이터 변수 저장
female = df_copy['sex'] == 'Female' # 성별이 'Female'인 조건 변수에 저장

df_copy.loc[spc_Chinstrap & female, 'bill_length_mm'].sort_values(ascending = False) # 두가지 조건을 모두 만족하는 데이터 들중에서 'bill_length_mm'열을 뽑고 값을 기준으로 정렬 하되 ascending을 False로 지정하여 내림 차순 정렬
```

```python
df_copy[spc_Chinstrap & female][['bill_length_mm']].sort_values(by = 'bill_length_mm', ascending = False)
```

```python
df_copy.head(2)
```

```python
# species(종)별 성별 구분하여 평균 몸무게 출력
# => ex) Adelie, Female / Adelie, male / Gentoo, Female...
df_copy['sex'].dropna()


spec = df_copy['species'].unique()
sex = df_copy['sex'].unique()

for i in spec:
    for j in sex[:2]:
        mean_weight = df_copy.loc[(df_copy['species']== i) & (df_copy['sex']== j), 'body_mass_g'].mean()
        print(f'{i}종의 성별 {j}의 평균 몸무게는 {mean_weight}입니다')
```

```python
# bill_depth 결측치 채우기
# => 같은종 같은섬의 평균으로
df_copy = df.copy()

# 2개의 결측치 존재 확인
df_copy['bill_length_mm'].isna().sum()

# 결측치를 갖고있는 데이터 확인
df_copy.loc[df_copy['bill_length_mm'].isna()]
```

### 'bill_depth'열에 결측치를 갖고있는 데이터의 종과 섬
1. species = Adelie, island = Torgersen
2. species = Gentoo, island = Biscoe

```python
# species = Adelie, island = Torgersen의 조건을 만족하는 데이터들의 'bill_depth' 평균 구하기
adelie_mean = df_copy.loc[(df_copy['species'] == 'Adelie') & (df_copy['island'] == 'Torgersen'), 'bill_depth_mm'].mean()
adelie_mean
```

```python
df_copy.loc[(df_copy['species'] == 'Adelie') & (df_copy['island'] == 'Torgersen'), 'bill_depth_mm'] = df_copy.loc[(df_copy['species'] == 'Adelie') & (df_copy['island'] == 'Torgersen'), 'bill_depth_mm'].fillna(adelie_mean)
```

```python
df_copy.head()
```

```python
# species = Gentoo, island = Biscoe의 조건을 만족하는 데이터들의 'bill_depth' 평균 구하기
Gentoo_mean = df_copy.loc[(df_copy['species'] == 'Gentoo') & (df_copy['island'] == 'Biscoe'), 'bill_depth_mm'].mean()
Gentoo_mean
```

```python
df_copy.loc[(df_copy['species'] == 'Gentoo') & (df_copy['island'] == 'Biscoe'), 'bill_depth_mm'] = df_copy.loc[(df_copy['species'] == 'Gentoo') & (df_copy['island'] == 'Biscoe'), 'bill_depth_mm'].fillna(adelie_mean)
```
