### practice

```python
tips = sns.load_dataset('tips')
tips.head()
```

- 남성 흡연자 조건 필터링
- tip을 가장 많이 낸 순서로 내림차순 정렬
- 상위 10개 출력

```python
# 남자 흡연자 조건 필터링

tips_male = tips['sex'] == 'Male'
tips_smoker = tips['smoker'] == 'Yes'
tips_male_smoker = tips.loc[tips_male & tips_smoker].head()
tips_male_smoker
## 남성 흡연자 들중 tip을 가장 많이 낸 순서내로 내림 차순 정렬
tips_male_ascdown = tips.loc[tips_male & tips_smoker].sort_values('tip', ascending=False)
tips_male_ascdown

## 남성 흡연자 들중 tip을 가장 많이 낸 순서내로 내림 차순 정렬 상위 10개
tips_male_ascdown.head(10)
```

- 식사 인원이 5명 이상인 저녁 테이블
- 흡연자가 없는 주말 테이블
- total_bil, tip 출력

```python
# 식사 인원이 5명 이상인 저녁 테이블
tips_dinner = tips['time'] == 'Dinner'
tips_size_5over = tips['size'] >= 5
tips_dinner_5 = tips.loc[tips_dinner & tips_size_5over]
tips_dinner_5
```

```python
# 흡연자가 없는 주말 테이블
tips_NoSmoker = tips['smoker'] == 'No'
tips_weekend = tips['day'].isin(['Sat','Sun'])

tips.loc[(tips_dinner & tips_size_5over) & (tips_NoSmoker & tips_weekend), ['total_bill', 'tip']]
```
