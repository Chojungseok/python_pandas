## 자전게 데이터 활용

```python
df = pd.read_csv('data/seoul_bicycle.csv')
df.head(2)
```

```python
df.info()
# 대여일자 열의 datatype이 object로 글자로 인식되고 있음
```

```python
df_copy = df.copy()
```

```python
# to_datetime(): 날짜 데이터로 바꾸어준다
df_copy['대여일자'] = pd.to_datetime(df_copy['대여일자'], format = '%b-%d-%Y')
```

```python
df_copy.info()
```

```python
cond = df_copy['대여일자'].dt.dayofweek == 3
df_copy.loc[cond]
```

```python
df_copy.head(3)
```

```python
pd.to_numeric(df_copy['운동량'])
```

```python
df_copy.loc[2344]
```

```python
# errors = '' : 에러가 났을때 어떻게 처리할 것인가 / 'coerce'은 nan으로 바꾸어 준다
df_copy['운동량'] = pd.to_numeric(df_copy['운동량'], errors = 'coerce')
```

```python
df_copy.info()
```

### cut()

- 내가 지정한 구간으로 나눠서 어디에 위치한지 알려준다

```python
bins = [0, 100, 200, 300, df_copy['운동량'].max()]
pd.cut(df_copy['운동량'], bins)
```

```python
pd.cut(df_copy['운동량'], bins = 10)
pd.cut(df_copy['운동량'], bins = 10).value_counts()
```

```python
## qcut() : 데이터를 일정한 비율로 나눈다
pd.qcut(df_copy['운동량'],q = 10)
pd.qcut(df_copy['운동량'],q = 10).value_counts()
```

```python
qcut_bins = [0, 0.2, 0.8, 1]
df_copy['운동량_한글'] = pd.qcut(df_copy['운동량'], qcut_bins, labels = ['적음', '보통', '많음'])
df_copy.head(2)
```
